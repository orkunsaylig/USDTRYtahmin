```
def otomatikMethod():
  for i in range(1,38,3):
    dosya = '/content/frames/'
    img = cv2.imread(dosya+ str(i) +'.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    ret,img = cv2.threshold(img,60,255,cv2.THRESH_BINARY)

    """
    """

    # çalışıyor
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(2,2))

    img_dilation = cv2.dilate(img, kernel, iterations = 2)


    """
    """

    # çalışıyor
    des = img_dilation
    contour,hier = cv2.findContours(des,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contour:
        cv2.drawContours(des,[cnt],0,255,-1)

    dilation = des


    """
    """
    # çalışıyor
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
    closing = cv2.morphologyEx(dilation, cv2.MORPH_CLOSE, kernel, iterations = 2)


    """
    """


    # çalışıyor
    des = closing
    contour,hier = cv2.findContours(des,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contour:
        cv2.drawContours(des,[cnt],0,255,-1)

    closing = des


    closing = cv2.bitwise_not(closing)

    cv2.imwrite('/content/otomatik/'+str(i)+'.jpeg',closing)

```

# Fonksiyonlar

```
def alan(dosya):
  image = cv2.imread(dosya)
  alan = 0
  beyaz = np.array((200,200,200))
  for  satir in image:
    for piksel in satir:
      sonuc = piksel >= beyaz
      if np.all(sonuc == True):
        alan += 1

  return alan
```

```
def doldur(image,sol,sag,yara):
  sol = np.array(sol)
  sag = np.array(sag)
  if yara != 0:
    yara = np.array(yara)
    image = cv2.fillPoly(image,[sol,sag,yara],0)
  else:
    image = cv2.fillPoly(image,[sol,sag],0)
  return image
```

```
def maskeOlustur(dosya_yolu):
  for i in range(1,36,3):
    sol, sag, yara = yaraNoktalari(dosya_yolu+'/'+str(i)+'.jpg.json')

    bos_image = np.ones((1440,1920), dtype = np.ubyte) * 255
    image = doldur(bos_image, sol, sag, yara)
    cv2.imwrite('/content/mask/'+str(i)+'.jpeg',image)

    
  for i in range(22,36,3):
    dosya = '/content/mask/'+str(i)+'.jpeg'

    im = Image.open(dosya).convert('RGB')
    im_invert = ImageOps.invert(im)
    im_invert.save(dosya)
```

```

def uzaklik(dosya):
  image = cv2.imread(dosya)
  uzakliklar = list()
  for satir in image:
    uzaklik = np.count_nonzero(satir == np.array([255,255,255]))
    uzakliklar.append(uzaklik)
  return np.array(uzakliklar)

```

```
def yaraNoktalari(dosya):
  yara = 0
  yara_iki = 0
  yara_uc = 0
  
  with open(dosya,'r') as file:
    veri = json.load(file)
    try:
      
      yara = (veri['objects'][0]['points']['exterior'])
      yara_iki = (veri['objects'][1]['points']['exterior'])
      yara_uc = (veri['objects'][2]['points']['exterior'])
      yara_dort =  (veri['objects'][3]['points']['exterior'])

    except IndexError:
      return yara,yara_iki,yara_uc
    
```

```
def cizgi(image,sol):
  try:
    
    for i,nokta in enumerate(sol):
      if i == 0:
        continue
      
      renk = (0,255,0)
      image = cv2.line(image,tuple(sol[i-1]),tuple(nokta),renk,thickness = 2)
    return image
  except :
    print('eyvah')

    
```

```
def alan(dosya):
  image = cv2.imread(dosya)

  unique, counts = np.unique(image, return_counts = True)

  area = dict(zip(unique,counts))

  beyaz_alan = area[255] / 3

  siyah_alan = (sum(area.values()) - beyaz_alan * 3) / 3

  return int(beyaz_alan), int(siyah_alan)

```
